# Language Increase Ratio

```python
worked_with_counts = survey_2023['LanguageHaveWorkedWith'].str.split(';').explode().value_counts()
want_to_work_with_counts = survey_2023['LanguageWantToWorkWith'].str.split(';').explode().value_counts()


merged_counts = pd.merge(worked_with_counts, want_to_work_with_counts, left_index=True, right_index=True, how='outer')
merged_counts.columns = ['Number of People Know', 'Number of People Want to Learn']

# NaN deÄŸerleri sÄ±fÄ±r ile dolduralÄ±m
merged_counts = merged_counts.fillna(0)

#popularity 
merged_counts['Increase_Ratio'] = (merged_counts['Number of People Want to Learn'] - merged_counts['Number of People Know'])/merged_counts['Number of People Know']


```

```python

merged_counts_sorted = merged_counts.sort_values(by='Increase_Ratio', ascending=False)
merged_counts_sorted.head(10)
```

```python
merged_counts_sorted.tail(10)
```

```python
# ArtÄ±ÅŸ oranÄ±na gÃ¶re sÄ±ralayalÄ±m
merged_counts_sorted = merged_counts.sort_values(by='Increase_Ratio', ascending=False)

# Bar plot
plt.figure(figsize=(10, 8))
merged_counts_sorted['Increase_Ratio'].plot(kind='bar', color='blue')
plt.title('Increase Ratio of Languages')
plt.xlabel('Language')
plt.ylabel('Increase Ratio')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
